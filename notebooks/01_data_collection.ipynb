```python

vocabulary_list = set()

for line in train['Tweets_split']:

  vocabulary_list.update(line)
len_vector = len(vocabulary_list)
print(len(vocabulary_list))
```

```python
#Index dictionary to assign an index to each word in vocabulary
from collections import defaultdict
index_word = {}
i = 0
for word in vocabulary_list:
    index_word[word] = i
    i += 1

print(index_word)
```

```python
train_bow = train['Tweet_without_stopwords'].head(10)
```

```python
def bag_of_words(sent):
    count_dict = defaultdict(int)
    vec = np.zeros(len_vector)
    #print(sent)
    for item in sent:
      if item in vocabulary_list:
       count_dict[item] += 1

      for key,item in count_dict.items():
          vec[index_word[key]] = item
    for i, num in enumerate(vec):
      vec[i] = num + 1
    return vec
```
